```python
import numpy as np
import matplotlib.pyplot as plt
from sys import argv
```

```python
def rpoint(x, y, r, deg):
    """ Get a point on a radius at deg degrees."""
    rang = np.deg2rad(deg)
    x1 = x + r * np.cos(rang)
    y1 = y + r * np.sin(rang)
    return (x1, y1)
```

```python
f = open("C:\\Users\\Gordon\\Documents\\Personal\\NeuralVroom\\pathfinder\\output.txt", "r")
parselist = [eval(line.rstrip('\n')) for line in f]
state_x = [pos[0] for pos in parselist]
state_y = [pos[1] for pos in parselist]
state_o = [pos[2] for pos in parselist]

start = (parselist[0][0], parselist[0][1], parselist[0][2])
end = (parselist[-1][0], parselist[-1][1], parselist[-1][2])
```

```python
fig, ax = plt.subplots()
ax.plot(start[0], start[1], marker='.', color='green', linestyle='none', MarkerSize=40)
x, y, a, b = start[0], start[1], 1, np.rad2deg(start[2])
x1, y1 = rpoint(x,y,a,b)
plt.arrow(x, y, (x1-x)/2, (y1-y)/2, head_width = .4, width = .1,color='green')
ax.plot(end[0], end[1], marker='.', color='red', linestyle='none', MarkerSize=40)
x, y, a, b = end[0], end[1], 1, np.rad2deg(end[2])
x1, y1 = rpoint(x,y,a,b)
plt.arrow(x, y, (x1-x)/2, (y1-y)/2, head_width = .4, width = .1,color='red')
ax.plot(state_x, state_y, marker='*', color='purple', MarkerSize=10)

# #draw gridlines
ax.grid(which='major', axis='both', linestyle='-', color='k', linewidth=1)
ax.set_xticks(np.arange(0, 40, 1))
ax.set_yticks(np.arange(0, 40, 1))
for index in range(len(state_o)):
    x, y, a, b = state_x[index], state_y[index], 1, np.rad2deg(state_o[index])
    x1, y1 = rpoint(x,y,a,b)
    plt.arrow(x, y, (x1-x)/2, (y1-y)/2, head_width = .5, width = .1,)
    
plt.rcParams['figure.figsize'] = [18, 18]
plt.show()
```
